```python
# Import of libraries needed
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression


# Data reading
# Since in this case I need to analize the data as different experiments, I can create two different df's
df1 = pd.read_csv("datos_pesas_r1.csv")
df2 = pd.read_csv("datos_pesas_r2.csv")
```

```python
# Now an errorbar must be done to not only show the measurements, but also the error margin for each measure done.
def errorbar_r1():

    x = df1["mass"]
    y = df1["longitude_r1"]
    xerr = df1["mass_error"]
    yerr = df1["longitude_error"]
    
    fig = plt.subplots()
    
    plt.errorbar(x, y, yerr, xerr)
    plt.xlabel('Mass (g)')
    plt.ylabel('Longitude (cm)')
    
    plt.savefig('pesas_r1.png')
    return fig

errorbar_r1()

# Errorbar for the second string
def errorbar_r2():

    x = df2["mass"]
    y = df2["longitude_r2"]
    xerr = df2["mass_error"]
    yerr = df2["longitude_error"]
    
    fig = plt.subplots()
    
    plt.errorbar(x, y, yerr, xerr)
    plt.xlabel('Mass (g)')
    plt.ylabel('Longitude (cm)')

    plt.savefig('pesas_r2.png')
    return fig

errorbar_r2()
```

```python
# Errorbar for both datasets

x1 = df1["mass"]
y1 = df1["longitude_r1"]
xerr1 = df1["mass_error"]
yerr1 = df1["longitude_error"]

x2 = df2["mass"]
y2 = df2["longitude_r2"]
xerr2 = df2["mass_error"]
yerr2 = df2["longitude_error"]

def errorbar_r1_r2():
    
    fig = plt.figure()
        
    plt.errorbar(x1, y1, xerr1, yerr1, ls='none')
    plt.errorbar(x2, y2, yerr2, xerr2, ls='none')
    plt.xlabel('Mass (g)')
    plt.ylabel('Longitude (cm)')
    
    plt.savefig('pesas_r1_r2.png')
    return fig

errorbar_r1_r2()
```

```python
# Linear regression for previous data
sns.regplot(x1, y1, ci=None)
sns.regplot(x2, y2, ci=None)
```
